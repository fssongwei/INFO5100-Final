```python
import pandas as pd
```

```python
dfHappy = pd.read_csv("happinessScore.csv")
dfAlcohol = pd.read_csv("HappinessAlcoholConsumption.csv")
```

```python
dfHappy.rename(columns={'Country Name': 'covid_count', 'COUNT': 'review_count'})
```

```python
dfResult = dfHappy.join(dfAlcohol.set_index('Country'), on="Country name", how='inner')
```

```python
dfResult = dfResult.drop(columns=['Regional indicator', 'Ladder score', 'Standard error of ladder score', 'upperwhisker', 
                       'lowerwhisker', 'Perceptions of corruption', 'Ladder score in Dystopia', 'Explained by: Log GDP per capita', 
                       'Explained by: Social support', 'Explained by: Healthy life expectancy', 'Explained by: Freedom to make life choices',
                       'Explained by: Generosity', 'Explained by: Perceptions of corruption', 'Dystopia + residual', 'HDI', 'Logged GDP per capita'])
```

```python
dfResult = dfResult.round(3)
```
